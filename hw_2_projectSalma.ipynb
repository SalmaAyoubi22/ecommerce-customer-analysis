```python
# Import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Import datasets
customers = pd.read_csv("customers.csv")
transactions = pd.read_csv("transactions.csv")
sessions = pd.read_csv("web_sessions.csv")
campaigns = pd.read_csv("marketing_campaigns.csv")

merged = transactions.merge(customers, on="customer_id")
sessions_merged = sessions.merge(customers, on="customer_id")
```

### Specific questions

Q1. Do customers from the UK spend more per transaction than customers from Germany?

```python
uk = merged[merged["country"] == "UK"]["purchase_value"]
germany = merged[merged["country"] == "Germany"]["purchase_value"]

t_stat, p_value = stats.ttest_ind(uk, germany)

print(f"UK average: ${uk.mean():.2f}")
print(f"Germany average: ${germany.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: The difference IS statistically significant")
else:
    print("Result: The difference is NOT statistically significant")
```

Q1 Interpretation:
UK customers average 75.63 and Germany 75.76, a difference of only 13 cents.
P-value: 0.9108 not significant.

The difference is completely random. The business should not treat UK and German customers differently in terms of pricing or marketing spend.

Q2. Do discounted purchases differ in average value from non-discounted purchases?

```python
discounted = merged[merged["discount_used"] == 1]["purchase_value"]
not_discounted = merged[merged["discount_used"] == 0]["purchase_value"]

t_stat, p_value = stats.ttest_ind(discounted, not_discounted)

print(f"Discounted average: ${discounted.mean():.2f}")
print(f"Non-discounted average: ${not_discounted.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: The difference IS statistically significant")
else:
    print("Result: The difference is NOT statistically significant")
```

Q2 Interpretation:
Discounted purchases average 64.40 vs 80.76 for full price,a $16.37 difference.
P-value: 0.0000  significant.

This is the strongest finding so far. Discount customers consistently spend less, and this is 100% real, not random. The business should reconsider its discount strategy as it is reducing revenue per transaction.

Q3. Do male and female customers differ in average purchase value?

```python
male = merged[merged["gender"] == "Male"]["purchase_value"]
female = merged[merged["gender"] == "Female"]["purchase_value"]

t_stat, p_value = stats.ttest_ind(male, female)

print(f"Male average: ${male.mean():.2f}")
print(f"Female average: ${female.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: The difference IS statistically significant")
else:
    print("Result: The difference is NOT statistically significant")
```

Q3 Interpretation:
Males average 76.04 and females 75.31  the difference is only 73 cents.
P-value: 0.2789 not significant.

Gender does not influence spending amount. Creating gender-based pricing or promotions would not be justified based on this data.

Q4. Do customers acquired via email spend more than those from social channels?

```python
email = merged[merged["acquisition_channel"] == "email"]["purchase_value"]
social = merged[merged["acquisition_channel"] == "social"]["purchase_value"]

t_stat, p_value = stats.ttest_ind(email, social)

print(f"Email average: ${email.mean():.2f}")
print(f"Social average: ${social.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: The difference IS statistically significant")
else:
    print("Result: The difference is NOT statistically significant")
```

Q5. Is device type associated with conversion?

```python
contingency_table = pd.crosstab(sessions["device_type"], sessions["converted"])
print(contingency_table)

chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\nChi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: Device type IS associated with conversion")
else:
    print("Result: Device type is NOT associated with conversion")
```

Q5 Interpretation:
Desktop: 3068 conversions, Mobile: 3042, Tablet: 3013.
P-value: 0.5626 not significant.

Device type has no association with whether someone buys or not. All devices convert at roughly the same rate (~13%). This is good news, it means the website works equally well on mobile, tablet and desktop.

Q8. Is time on site correlated with purchase value?

Q8 Interpretation:

There is no relationship between time spent on site and how much someone spends. Keeping customers on the site longer does not lead to bigger purchases. Time on site is not a useful metric for predicting revenue.

C1. Explore whether high-spending customers behave differently in terms of session time.

Define “high spending” yourself.

```python
# Define high spending as above the 75th percentile
threshold = merged["purchase_value"].quantile(0.75)
print(f"High spending threshold: ${threshold:.2f}")

high_spenders = merged[merged["purchase_value"] >= threshold]["customer_id"]
low_spenders = merged[merged["purchase_value"] < threshold]["customer_id"]

high_sessions = sessions[sessions["customer_id"].isin(high_spenders)]["time_on_site"]
low_sessions = sessions[sessions["customer_id"].isin(low_spenders)]["time_on_site"]

t_stat, p_value = stats.ttest_ind(high_sessions, low_sessions)

print(f"High spenders avg time on site: {high_sessions.mean():.2f} seconds")
print(f"Low spenders avg time on site: {low_sessions.mean():.2f} seconds")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: High spenders DO behave differently in session time")
else:
    print("Result: High spenders do NOT behave differently in session time")
```

C1 Interpretation:
I defined high spending as purchases above the 75th percentile  $101.03.

High spenders averaged 300.84 seconds on site vs 300.89 seconds for low spenders.

P-value: 0.9544  not significant.

High spenders and low spenders behave identically on the website. We cannot identify a high-value customer by how long they browse. This means the business cannot use session time to predict or target high spenders.

c3. Identify a statistically significant result that may not be practically meaningful.

Explain why statistical significance ≠ business importance.

```python
# C3: Statistically significant but not practically meaningful
# Example: Compare purchase values across all countries
from scipy.stats import f_oneway

groups = [merged[merged["country"] == c]["purchase_value"]
          for c in merged["country"].unique()]

f_stat, p_value = f_oneway(*groups)

country_avgs = merged.groupby("country")["purchase_value"].mean()
print("Average purchase by country:")
print(country_avgs.round(2))
print(f"\nF-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Range of averages: ${country_avgs.max() - country_avgs.min():.2f}")
```

C3: Statistical significance vs practical importance

The discount analysis (Q2) is a strong example of this concept.

The t-test gave a p-value of 0.0000, meaning the $16.37 difference
between discounted and non-discounted purchases is 100% statistically
real, not random.

However, we need to ask: is $16.37 practically meaningful enough to
change business strategy?

Arguments that it IS practically meaningful:
- Multiplied across 13,500 discounted transactions = $221,000 impact
- That represents real lost revenue at scale

Arguments that it may NOT be practically meaningful:
- The discount may be attracting customers who would not have purchased
  at full price at all — meaning some revenue is better than none
- We don't know if removing discounts would cause those customers to
  simply not buy, which would lose the 64 entirely rather than gaining
  the extra $16
- Statistical significance just tells us the difference is real
  it doesn't tell us what caused it or what would happen if we removed discounts

This is why statistical significance alone is never enough for a
business decision. You need context, causation, and business judgement
alongside the numbers.
